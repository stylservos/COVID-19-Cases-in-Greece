```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import *
import sys    
    
def create_visualizations(filename):
    df = pd.read_csv(filename, sep=',', encoding="utf-8", engine='python')
    df["Year"] = df.Date.str.split('-', expand = True)[0]
    df["Month"] = df.Date.str.split('-', expand = True)[1]
    df["Day"] = df.Date.str.split('-', expand = True)[2]
    df['Year_Int'] = df['Year'].fillna(0).astype(int)
    df = df.loc[df['Year_Int'] >= 2020]
    df = df.groupby(['Month', 'Year', 'Area'])['Cases'].sum().reset_index(name='Cases')
    df['Month_Int'] = df['Month'].fillna(0).astype(int)
    min_month = df['Month_Int'].min()
    max_month = df['Month_Int'].max()
    months = [min_month, max_month]
    areas = [u'Όλα'] + list(df['Area'].unique())
    years = [u'Όλα'] + list(df['Year'].unique())
    df2 = df.groupby(['Month', 'Year'])['Cases'].sum().reset_index(name='Cases')
    df2['Month_Int'] = df2['Month'].fillna(0).astype(int)
    df3 = df.groupby(['Year', 'Area'])['Cases'].sum().reset_index(name='Cases')
    df4 = df.groupby(['Year'])['Cases'].sum().reset_index(name='Cases')
    df5 = df.groupby(['Month', 'Area'])['Cases'].sum().reset_index(name='Cases')
    df5['Month_Int'] = df5['Month'].fillna(0).astype(int)
    df6 = df.groupby(['Month'])['Cases'].sum().reset_index(name='Cases')
    df6['Month_Int'] = df6['Month'].fillna(0).astype(int)
    
    def plot_bar(month, year):
        data = df
        data2 = df5
        if (year != u'Όλα'):
            df_filtered = data.loc[(data['Month_Int'] == month) & (data['Year'] == year)]
            plt.figure(1, figsize=(14, 6))
            plt.bar(df_filtered['Area'], df_filtered['Cases'])
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Γεωγραφικό Διαμέρισμα", fontsize=16)
            plt.xlabel(u"Γεωγραφικό Διαμέρισμα", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16)
        else:
            df_filtered = data2.loc[data2['Month_Int'] == month]
            plt.figure(1, figsize=(14, 6))
            plt.bar(df_filtered['Area'], df_filtered['Cases'])
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Γεωγραφικό Διαμέρισμα", fontsize=16)
            plt.xlabel(u"Γεωγραφικό Διαμέρισμα", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16)
            
    def plot_line(area, months, year):
        data = df
        data2 = df2
        data3 = df5
        data4 = df6
        if (area != u'Όλα' and year != u'Όλα'):
            df_filtered = data.loc[(data['Area'] == area) & (data['Year'] == year) & (data['Month_Int'] >= months[0]) & (data['Month_Int'] <= months[1])]
            plt.figure(1, figsize=(14, 6))
            plt.plot(df_filtered['Month'], df_filtered['Cases'])
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Μήνα", fontsize=16)
            plt.xlabel(u"Μήνας", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16) 
        elif (area == u'Όλα' and year != u'Όλα'):
            df_filtered = data2.loc[(data2['Year'] == year) & (data2['Month_Int'] >= months[0]) & (data2['Month_Int'] <= months[1])]
            plt.figure(1, figsize=(14, 6))
            plt.plot(df_filtered['Month'], df_filtered['Cases'])
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Μήνα", fontsize=16)
            plt.xlabel(u"Μήνας", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16) 
        elif (area != u'Όλα' and year == u'Όλα'):
            df_filtered = data3.loc[(data3['Area'] == area) & (data3['Month_Int'] >= months[0]) & (data3['Month_Int'] <= months[1])]
            plt.figure(1, figsize=(14, 6))
            plt.plot(df_filtered['Month'], df_filtered['Cases'])
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Μήνα", fontsize=16)
            plt.xlabel(u"Μήνας", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16)
        elif (area == u'Όλα' and year == u'Όλα'):
            df_filtered = data4.loc[(data4['Month_Int'] >= months[0]) & (data4['Month_Int'] <= months[1])]
            plt.figure(1, figsize=(14, 6))
            plt.plot(df_filtered['Month'], df_filtered['Cases'])
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Μήνα", fontsize=16)
            plt.xlabel(u"Μήνας", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16) 
                
    def plot_area(area):
        data = df3
        data2 = df4
        if (area != u'Όλα'):
            df_filtered = data.loc[data['Area'] == area]
            plt.figure(1, figsize=(14, 6))
            plt.fill_between(list(df_filtered['Year']), list(df_filtered['Cases']),
              color="skyblue", alpha=0.4)
            plt.plot(list(df_filtered['Year']), list(df_filtered['Cases']), color="Slateblue",
                     alpha=0.6, linewidth=2)
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Έτος", fontsize=16)
            plt.xlabel(u"Έτος", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16)
        else:
            df_filtered = data2
            plt.figure(1, figsize=(14, 6))
            plt.fill_between(list(df_filtered['Year']), list(df_filtered['Cases']),
              color="skyblue", alpha=0.4)
            plt.plot(list(df_filtered['Year']), list(df_filtered['Cases']), color="Slateblue",
                     alpha=0.6, linewidth=2)
            plt.title(u"Άθροισμα κρουσμάτων COVID-19 στην Ελλάδα ανά Έτος", fontsize=16)
            plt.xlabel(u"Έτος", fontsize=16)
            plt.ylabel(u"Κρούσματα COVID-19", fontsize=16)
    month = widgets.IntSlider(value=min_month,min=min_month,max=max_month,step=1,description=u'Μήνας')
    year = widgets.Dropdown(options=years, value=u'Όλα', description=u'Έτος')
    area = widgets.Dropdown(options=areas,value=u'Όλα',description=u'Διαμέρισμα')
    months = widgets.IntRangeSlider(value=months,min=min_month,max=max_month,step=1,description=u'Μήνες')
        
    month.set_title = 'month'
    year.set_title = 'year'
    area.set_title = 'area'
    months.set_title = 'months'
          
    #assign widgets to tabs
    tab1 = widgets.HBox([month, year])
    tab2 = widgets.HBox([area, months, year])
    tab3 = widgets.HBox([area])

    #create tabs
    tab_nest = widgets.Tab()
    tab_nest.set_title(0, u'Κρούσματα ανά Διαμέρισμα')
    tab_nest.set_title(1, u'Κρούσματα ανά Μήνα')
    tab_nest.set_title(2, u'Κρούσματα ανά Έτος')

    #interact function in isolation
    f = interactive(plot_bar, month = month, year = year);
    f2 = interactive(plot_line, area = area, months = months, year = year);
    f3 = interactive(plot_area, area = area);
    tab_nest.children = [VBox(children = f.children), VBox(children = f2.children), VBox(children = f3.children)]
    display(tab_nest)  
    
def main():
    reload(sys)
    sys.setdefaultencoding("utf-8")
    pathname = "Datasets/new_COVID-19_Cases_Greece.csv"
    create_visualizations(pathname)

main()
```
